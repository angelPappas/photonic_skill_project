Calculate the following questions using femwell mode solver:

● Calculate the refractive index change due to a reverse biased PN junction. Calculate the
length needed for a PI phase shift as well as the insertion loss

```python
from collections import OrderedDict

import matplotlib.pyplot as plt
import numpy as np
import shapely
from skfem import Basis, ElementTriP0
from skfem.io.meshio import from_meshio

from femwell.maxwell.waveguide import compute_modes
from femwell.mesh import mesh_from_OrderedDict
from femwell.pn_analytical import index_pn_junction, k_to_alpha_dB
```

```python
clad_thickness = 2
slab_width = 3
wg_width = 0.5
wg_thickness = 0.22
slab_thickness = 0.09

core = shapely.geometry.box(-wg_width / 2, -wg_thickness / 2, wg_width / 2, wg_thickness / 2)
slab = shapely.geometry.box(
    -slab_width / 2, -wg_thickness / 2, slab_width / 2, -wg_thickness / 2 + slab_thickness
)
clad = shapely.geometry.box(
    -slab_width / 2, -clad_thickness / 2, slab_width / 2, clad_thickness / 2
)

polygons = OrderedDict(
    core=core,
    slab=slab,
    clad=clad,
)

resolutions = dict(
    core={"resolution": 0.02, "distance": 0.5}, slab={"resolution": 0.04, "distance": 0.5}
)

mesh = from_meshio(mesh_from_OrderedDict(polygons, resolutions, default_resolution_max=10))
mesh.draw().show()
```

```python
def define_index(mesh, V, xpn, NA, ND, wavelength) -> tuple[Basis, np.ndarray]:
    """
    Builds the complex permitivity distribution on the FEM mesh:
    1. Start with silicon refractive index in the core and slab
    2. Add the complex index perturbation from the p-n junction (using the analytical model from Soref and Bennett, femwell.pn_analytical.index_pn_junction)
    3. Overwrite the cladding region with SiO2 refractive index
    4. Square the refractive index to get the permitivity, which is the quantity used in the FEM solver

    Parameters
    ----------
    mesh: skfem Mesh
        The FEM mesh on which to define the index distribution
    V: float
        The applied voltage across the p-n junction (in Volts)
    xpn: float
        The lateral position of the p-n junction boundary (in microns)
    NA: float
        The acceptor doping concentration (in cm^-3)
    ND: float
        The donor doping concentration (in cm^-3)
    wavelength: float
        The operating wavelength (in microns)
    
    Returns
    -------
    basis0: skfem Basis
        The FEM basis on which the index distribution is defined
    epsilon: np.ndarray
        The complex permitivity distribution defined on the FEM basis
    """
    basis0 = Basis(mesh, ElementTriP0())
    epsilon = basis0.zeros(dtype=complex)
    for subdomain, n in {"core": 3.45, "slab": 3.45}.items():
        epsilon[basis0.get_dofs(elements=subdomain)] = n
    epsilon += basis0.project(
        lambda x: index_pn_junction(x[0], xpn, NA, ND, V, wavelength),
        dtype=complex,
    )
    for subdomain, n in {"clad": 1.444}.items():
        epsilon[basis0.get_dofs(elements=subdomain)] = n
    epsilon *= epsilon  # square now
    return basis0, epsilon
```

```python
xpn = 0 # lateral position of the p-n junction boundary [um]
NA = 1e18   # acceptor doping concentration [cm^-3]
ND = 1e18   # donor doping concentration [cm^-3]
wavelength = 1.55

V = -4
basis0, epsilon = define_index(mesh, V, xpn, NA, ND, wavelength)
basis0.plot(epsilon.imag, colorbar=True).show()
```

```python
voltages = [0, -1, -2, -3, -4]
neff_vs_V = []
for V in voltages:
    basis0, epsilon = define_index(mesh, V, xpn, NA, ND, wavelength)
    modes = compute_modes(basis0, epsilon, wavelength=wavelength, num_modes=1, order=2)
    neff_vs_V.append(modes[0].n_eff)
```

### Results

#### 1. Refractive index change

```python
plt.plot(voltages, np.real(neff_vs_V) - np.real(neff_vs_V[0]), marker="o", label="Re[neff]")
plt.title(f"NA = {NA}, ND = {ND}, xpn = {xpn}, wavelength = {wavelength}", fontsize=10)
plt.xlabel("Voltage [V]", fontsize=10)
plt.ylabel("Change in neff", fontsize=10)
plt.grid()
plt.legend()
plt.show()
```

#### 2. L_pi calculation

```python
L_pi = wavelength*1e-6/(2* (np.real(neff_vs_V[2:]) - np.real(neff_vs_V[0])))    # in meters

for i,L in enumerate(L_pi):
    print(f"L_pi = {L*1e6:.2f} um, for V = {voltages[i+2]} V reverse bias")
```

Depending on the applied voltage:

- L_pi = 6468.38 um, for V = -2 V reverse bias
- L_pi = 4707.12 um, for V = -3 V reverse bias
- L_pi = 3775.33 um, for V = -4 V reverse bias

#### 3. Insertion Loss

```python
plt.plot(voltages, k_to_alpha_dB(np.imag(neff_vs_V), wavelength), marker="o", label="Loss")
plt.title(f"NA = {NA}, ND = {ND}, xpn = {xpn}, wavelength = {wavelength}", fontsize=10)
plt.xlabel("Voltage [V]", fontsize=10)
plt.ylabel("Loss [dB/cm]", fontsize=10)
plt.legend()
plt.grid()
plt.show()
```
